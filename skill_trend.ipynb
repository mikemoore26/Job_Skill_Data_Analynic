# How arre in-demand skills trending for Data Analysts>

Methology 

    1. Aggregate skill counts monthly

    2. Re-analyze based on percentafe of total jobs
    
    3. Plots the monthly skill demand

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
from ast import literal_eval

# Loading Data
data= load_dataset("lukebarousse/data_jobs")
df = data['train'].to_pandas()

#Data Clean Up
df["job_posted_date"] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: literal_eval(x) if not pd.isna(x) else x)
```

```python
df_us_da = df[(df["job_title"] == "Data Analyst") & (df['job_country'] == 'United States')].copy()
df_us_da.sample(2)
```

```python
df_us_da['job_posted_month_no'] = df_us_da['job_posted_date'].dt.month
```

```python
df_explode = df_us_da.explode("job_skills").copy()
df_explode.sample(3)
```

```python
df_pivot = df_explode.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)
df_pivot.loc['total'] = df_pivot.sum()
df_pivot = df_pivot[df_pivot.loc['total'].sort_values(ascending=False).index]
df_pivot.drop("total",inplace=True)
df_pivot
```

```python
#df_pivot.plot(kind="line")
```

```python
da_totals = df_us_da.groupby("job_posted_month_no").size()
da_totals
```

```python
df_perc = df_pivot.div(da_totals/100,axis=0)
df_perc
```

```python
df_perc = df_perc.reset_index()
df_perc['job_posted_month'] = df_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format="%m").strftime('%b'))
df_perc = df_perc.set_index("job_posted_month")
df_perc = df_perc.drop("job_posted_month_no", axis=1)
df_perc
```

```python
df_plot = df_perc.iloc[:, :5]
sns.set_theme(style="ticks")

sns.lineplot(df_plot, dashes=False, palette="tab10")
plt.title("Trending top skills for data analysis in the us")
plt.ylabel("Likihood in job posting")
plt.xlabel("2023")
plt.legend().remove()
sns.despine()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
